```python
import pandas as pd
import numpy as np

# assuming dfA and dfB are your dataframes
dfA = pd.DataFrame({'A': [1, 2, 3, 4, 5]}, index = ['a', 'b', 'c', 'd', 'e'])
dfB = pd.DataFrame({'B': ['a', 'a', 'b', 'c', 'c', 'c', 'd', 'd', 'd', 'd']})

# flatten dfB into a Series
dfB_flattened = pd.Series(dfB.values.ravel('K'))

# Count how many times each index of dfA occurs in dfB
index_counts = dfB_flattened.value_counts()
print(index_counts)
# Map the counts onto dfA to create a new column, filling NaN values with 0
dfA['counts'] = dfA.index.map(index_counts).fillna(0).astype(int)

print(dfA)
```
